## Reporting: wragle_report

### Introduction

This project was about tweets of dogs. The dataset is  the tweet archive of Twitter user @dog_rates, also known as WeRateDogs. WeRateDogs is a Twitter account that rates people's dogs with a humorous comment about the dog. 

In order to make the project, I had to pass through these phases:

Step 1: Gathering data

Step 2: Assessing data

Step 3: Cleaning data

Step 4: Storing data

Step 5: Analyzing, and visualizing data

Step 6: Reporting


### Gathering Data 

For gathering the data, I had to import the data from three sources.
1. WeRateDogs gave us an archive called 'twitter-archiveenhanced.csv'
2. Each Tweet had a prediction and this file was downloaded programatically using the URL
3. I used the Tweepy Library to download information about each tweet using the API of Twiter.

### Assessing  Data
I open each dataframe in order to watch what kind of information I had. I identify some of the Tidiness issues and the quality of them.

#### Quality issues
**df_archive**

- Transform the 'None' to empty cells of [doggo,floofer,pupper,puppo] and then unify in a 'dog class' column

- Transform the 'timestamp' column in a Date column

- The retweets can duplicate information

- Delete columns that will not help the analysis

**image_prediction**

- The column names such as p1,p2 are not descriptive.

- Duplicated values in the image_prediction_clean

- Dog names format should be consistent.

**tweet_status**

- Remove the retweets

#### Tidiness issues

The 'timestamap' column is a string it can be changed to a date format.

The dog classification can be in one column.
